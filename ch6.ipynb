## 处理xml文件

```python
from lxml.html import parse
from urllib2 import urlopen
```

```python
from lxml import objectify
path = 'G:/dataAnalysis/data/xml/Performance_MNR.xml'
parsed = objectify.parse(open(path))
root = parsed.getroot()
```

```python
root
```

```python
!more +1 G:\\dataAnalysis\\data\\xml\\Performance_MNR.xml
```

```python
data = []

skip_fields = ['PARENT_SEQ', 'INDICATOR_SEQ',
               'DESIRED_CHANGE', 'DECIMAL_PLACES']

for elt in root.INDICATOR:
    el_data = {}
    for child in elt.getchildren():
        if child.tag in skip_fields:
            continue
        el_data[child.tag] = child.pyval
    data.append(el_data)
    
df = DataFrame(data)
df
```

## 操作Microsoft Excel 文件

```python
import xlrd, xlwt
```

```python
excel_path = 'G:\dataAnalysis\data'
```

```python
wb = xlwt.Workbook()
wb.add_sheet('sheet1', cell_overwrite_ok=True)
wb.get_active_sheet()
```

```python
ws_1 = wb.get_sheet(0)
ws_1
```

```python
excel_data = np.arange(1,65).reshape((8,8))
excel_data
```

```python
for row in range(excel_data.shape[0]):
    for col in range(excel_data.shape[1]):
        ws_1.write(row, col, excel_data[row, col])

wb.save(excel_path+'\workbook.xls')
```
