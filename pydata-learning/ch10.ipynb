###  带有重复索引的时间序列

```python
dates = pd.DatetimeIndex(['1/1/2019', '1/2/2019', '1/2/2019', '1/2/2019',
                          '1/3/2019'])
dup_ts = Series(np.arange(5), index=dates)
dup_ts
```

```python
dup_ts.index.is_unique
```

```python
grouped = dup_ts.groupby(level=0)    #level=0 索引的唯一一层
grouped
```

```python
grouped.mean()
```

```python
grouped.count()
```

### 时区处理

```python
import pytz
```

```python
pytz.common_timezones[:8]
```

```python
[x for x in pytz.common_timezones if x.find('Shanghai') != -1]
```

```python
tz = pytz.timezone('Asia/Shanghai')
tz
```

```python
rng = pd.date_range('3/9/2012 9:30', periods=6, freq='D')
ts = Series(np.random.randn(len(rng)), index=rng)
```

```python
print ts.index.tz
```

```python
pd.date_range('3/9/2012 9:30', periods=10, freq='D', tz='UTC')
ts_utc = ts.tz_localize('UTC')     #时间序列本地化为某个具体的时区
ts_utc
```

```python
ts_utc.index
```

```python
ts_utc.tz_convert('US/Eastern')
```

```python
ts_eastern = ts.tz_localize('US/Eastern')
ts_eastern.tz_convert('UTC')
```

```python
ts_eastern.tz_convert('Asia/Shanghai')              #美国东部时间转为上海时间
```

```python
ts_eastern.index
```

```python
ts.index.tz_localize('Asia/Shanghai')
```
