## 数据库风格的DataFrame 合并

```python
df1 = DataFrame({'key':['b','b','a','c','a','a','b'], 'data1':np.arange(7)})
df2 = DataFrame({'key':['a','b','c'], 'data2':np.arange(3)})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

显式指定合并时用于连接的列， merge默认将重叠列的列名当做键

```python
pd.merge(df1, df2, on='key')
```

merge默认是作 "inner"连接

```python
df3 = DataFrame({'lkey':['b','b','a','c','a','a','b'], 'data1':np.arange(7)})
df4 = DataFrame({'rkey':['a','b','d'], 'data2':np.arange(3)})
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
df3
```

```python
df4
```

指定使用外连接

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, on= 'key', how='left')
```

```python
df5 = DataFrame({'key':['b','b','a','c','a','b'], 'data1':np.arange(6)})
df6 = DataFrame({'key':['a','b','a','b','d'], 'data2':np.arange(5)})
```

```python
df5
```

```python
df6
```

```python
pd.merge(df5, df6, on= 'key', how='left')
```

```python
pd.merge(df5, df6, on= 'key', how='inner')
```

```python
left = DataFrame({'key1': ['foo', 'foo', 'bar'],
                  'key2': ['one', 'two', 'one'],
                  'lval': [1, 2, 3]})
right = DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                   'key2': ['one', 'one', 'one', 'two'],
                   'rval': [4, 5, 6, 7]})
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
pd.merge(left, right, on='key1')
```

处理列名重复问题，使用 suffixes 选项

```python
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))
```

显然，没有prefixes 选项

```python
pd.merge(left, right, on='key1', prefixes=('left_', 'right_'))
```

## 索引上的合并

```python
left1 = DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                  'value': np.arange(6)})
right1 = DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
pd.merge(left1, right1, left_on='key', right_on='group_val')
```

默认的merge方法是求取连接键的交集，若想获取并集，则应使用外连接的方式

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

## 利用函数或映射进行数据转换

```python
data = DataFrame({'food': ['bacon', 'pulled pork', 'bacon', 'Pastrami',
                           'corned beef', 'Bacon', 'pastrami', 'honey ham',
                           'nova lox'],
                  'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

map() 方法接受一个函数或者含有映射关系的字典型对象

```python
meat_to_animal = {
  'bacon': 'pig',
  'pulled pork': 'pig',
  'pastrami': 'cow',
  'corned beef': 'cow',
  'honey ham': 'pig',
  'nova lox': 'salmon'
}
data['animal'] = data['food'].map(str.lower).map(meat_to_animal)
```

```python
data
```

```python
data['animal'] = data['food'].map(lambda x : meat_to_animal[x.lower()])
data
```

## 离散化和面元划分

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 62, 45, 43, 32]
bins = [18, 25, 35, 60 ,100]
cats = pd.cut(ages, bins)
cats
```

```python
type(cats)
```

```python
cats.labels
```

```python
# 上述异常表明属性名称的更新  categories 替换levels
cats.categories
```

```python
cats.codes     # codes 替换 labels
```

```python
pd.value_counts(cats)
```

```python
# 自己设置面元名称
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
pd.cut(ages, bins, labels=group_names)
```

```python
data = np.random.rand(20)
```

```python
data
```

```python
pd.cut(data, 4, precision=2)     #传入面元数量则根据数据的最小和最大值等值计算等长面元
```

```python
data = np.random.rand(1000)
data
```

```python
cats = pd.qcut(data, 4)
cats
```

```python
pd.value_counts(cats)     # 使用qcut() 可以得到大小基本相等的面元
```
